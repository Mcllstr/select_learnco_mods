```python
df.age.hist()  # Comment - age of homes is heavily skewed toward older units in data set, not surprising for Boston area.  
plt.title('distribution of age of homes')
plt.xlabel('proportion of owner occupied units built priot to 1940')  
```

```python
df.medv.hist()  # Comment - median home value appears close to normal.   
plt.title('distribution of median value of homes')
plt.xlabel('median home value')  
plt.ylabel('frequency')
```

```python
df.nox.hist()  # Comment - NOx concentrations have heavy right skew  
plt.title('NOx concentrations')
plt.xlabel('NOx ppm')  
plt.ylabel('frequency')
```

```python
grouped = df.groupby('nox').indus.median()

grouped.plot(kind = 'barh', figsize = (12,15))

plt.ylabel('NOx concentrations (ppm)')
plt.xlabel('median proportion non retail business acres per town')

plt.title('median industrial use proportions by NOx concentration')

# There appears to be correlation between high NOx concentrations and industrial space
```

```python
grouped = df.groupby('nox').age.median()

grouped.plot(kind = 'barh', figsize = (12,15))

plt.ylabel('NOx concentrations (ppm)')
plt.xlabel('median proportion bldgs over 40 yrs old')

plt.title('average age by NOx concentration')

#clear correlation between age and NOx concentrations
```

```python
grouped = df.groupby('indus').age.median()  ## Appears to be some correlation between age and industrial use
grouped
grouped.plot(kind = 'barh', figsize = (12,15))

plt.ylabel('median proportion non retail business acres per town')
plt.xlabel('Median proportion homes over 40 yrs old')

plt.title('average age by NOx concentration')
```

## Variable Descriptions

This data frame contains the following columns:

#### crim  
per capita crime rate by town.

#### zn  
proportion of residential land zoned for lots over 25,000 sq.ft.

#### indus  
proportion of non-retail business acres per town.

#### chas  
Charles River dummy variable (= 1 if tract bounds river; 0 otherwise).

#### nox  
nitrogen oxides concentration (parts per 10 million).

#### rm  
average number of rooms per dwelling.

#### age  
proportion of owner-occupied units built prior to 1940.

#### dis  
weighted mean of distances to five Boston employment centers.

#### rad  
index of accessibility to radial highways.

#### tax  
full-value property-tax rate per $10,000.

#### ptratio  
pupil-teacher ratio by town.

#### black  
1000(Bk - 0.63)^2 where Bk is the proportion of blacks by town.

#### lstat  
lower status of the population (percent).

#### medv  
median value of owner-occupied homes in $10000s.
